## 01_Bảng Chi nhánh

```python
import random
from datetime import datetime

SO_CHINHANH = 10

def fmt_cn(i):
    """Định dạng chuẩn CN01, CN02..."""
    return f"CN{str(i).zfill(2)}"

ten_chi_nhanh = [
    "Chi nhánh Quận 1", "Chi nhánh Quận 2", "Chi nhánh Quận 3",
    "Chi nhánh Quận 4", "Chi nhánh Quận 5", "Chi nhánh Quận 7",
    "Chi nhánh Tân Bình", "Chi nhánh Bình Thạnh",
    "Chi nhánh Gò Vấp", "Chi nhánh Phú Nhuận"
]

dia_chi_mau = [
    "12 Lê Lợi, Quận 1, TP.HCM", "45 Nguyễn Huệ, Quận 1, TP.HCM",
    "77 Lý Chính Thắng, Quận 3, TP.HCM", "23 Hoàng Sa, Quận 1, TP.HCM",
    "99 Võ Văn Tần, Quận 3, TP.HCM", "150 Trần Não, TP. Thủ Đức",
    "320 Cộng Hòa, Tân Bình, TP.HCM", "55 Điện Biên Phủ, Bình Thạnh, TP.HCM",
    "12 Quang Trung, Gò Vấp, TP.HCM", "88 Phan Xích Long, Phú Nhuận, TP.HCM"
]

def rand_time_open_close():
    open_hour = random.randint(7, 9)   # 7h-9h
    close_hour = random.randint(18, 22) # 18h-22h
    return (
        f"{open_hour:02d}:00",
        f"{close_hour:02d}:00"
    )

def rand_phone():
    prefixes = ["090", "091", "098", "097", "088", "035"]
    suffix = "".join(str(random.randint(0, 9)) for _ in range(7))
    return random.choice(prefixes) + suffix

with open("insert_chinhanh.sql", "w", encoding="utf-8") as f:
    sql_lines = []

    sql_lines.append("USE PetCareDB\nGO\n")
    sql_lines.append("DELETE FROM CHINHANH;\n")

    for i in range(1, SO_CHINHANH + 1):
        ma_cn = fmt_cn(i)
        ten = ten_chi_nhanh[i - 1]
        diachi = dia_chi_mau[i - 1]
        sdt = rand_phone()
        time_open, time_close = rand_time_open_close()

        sql = (
            f"INSERT INTO CHINHANH (MaCN, TenCN, TimeMoCua, TimeDongCua, DiaChiCN, SDT_CN) "
            f"VALUES (N'{ma_cn}', N'{ten}', '{time_open}', '{time_close}', N'{diachi}', '{sdt}');\n"
        )

        sql_lines.append(sql)

    f.write("".join(sql_lines))

print("ĐÃ TẠO FILE insert_chinhanh.sql (10 dòng) - Đã sửa thứ tự cột.")
```

## 04_Bảng Thuốc, sản phẩm

```python
import random
from datetime import datetime, timedelta

NUM_SANPHAM = 500
NUM_THUOC = 100

def fmt(prefix, i, length=5):
    return f"{prefix}{str(i).zfill(length)}"

def rand_future_date():
    """Hạn sử dụng thuốc: 2026 - 2028"""
    start = datetime(2026, 1, 1)
    end = datetime(2028, 12, 31)
    delta = end - start
    return (start + timedelta(days=random.randint(0, delta.days))).strftime("%Y-%m-%d")

random.seed(42)

print("--- BƯỚC 1: TẠO DỮ LIỆU ---")

sanpham_sql = []
thuoc_sql = []

thuoc_units = ["Viên", "Hộp", "Chai", "Vỉ", "Tuýp", "Gói"]
sp_types = ["Thức ăn", "Đồ chơi", "Phụ kiện", "Chuồng", "Vệ sinh", "Dinh dưỡng"]
sp_names = ["Hạt Royal Canin", "Pate Whiskas", "Bóng cao su", "Dây dắt", "Nhà cây", "Cát vệ sinh", "Sữa tắm", "Lược chải lông"]

all_sp_ids = [fmt("SP", i, 5) for i in range(1, NUM_SANPHAM + 1)]
random.shuffle(all_sp_ids)

thuoc_ids = all_sp_ids[:NUM_THUOC]
other_sp_ids = all_sp_ids[NUM_THUOC:]

for i, sp_id in enumerate(thuoc_ids):
    ten_thuoc = f"Thuốc {random.choice(['Kháng sinh', 'Giảm đau', 'Vitamin', 'Trị ve'])} {chr(65 + i%26)}{i}"
    don_vi_constraint = ["ống", "viên", "hộp", "mg", "g"]
    don_vi = random.choice(don_vi_constraint)

    hsd = rand_future_date()
    gia = random.randint(5, 100) * 5000

    sanpham_sql.append(
        f"INSERT INTO SANPHAM (MaSP, TenSP, LoaiSP, GiaBan) VALUES (N'{sp_id}', N'{ten_thuoc}', N'Thuốc', {gia});\n"
    )

    thuoc_sql.append(
        f"INSERT INTO THUOC (MaThuoc, DonViTinh, NgayHetHan) VALUES (N'{sp_id}', N'{don_vi}', '{hsd}');\n"
    )

for i, sp_id in enumerate(other_sp_ids):
    raw_type = random.choice(sp_types)

    if raw_type == "Thức ăn":
        loai_chuan = "Thức ăn"
    else:
        loai_chuan = "Phụ kiện"

    ten_sp = f"{random.choice(sp_names)} {random.choice(['Premium', 'Classic', 'Gold'])} {i}"
    gia = random.randint(2, 500) * 10000

    sanpham_sql.append(
        f"INSERT INTO SANPHAM (MaSP, TenSP, LoaiSP, GiaBan) VALUES (N'{sp_id}', N'{ten_sp}', N'{loai_chuan}', {gia});\n"
    )

print("--- BƯỚC 2: GHI FILE SQL VỚI 'NOCHECK CONSTRAINT' ---")

with open("insert_master_products.sql", "w", encoding="utf-8-sig") as f:
    f.write("USE PetCareDB\nGO\n")

    f.write("-- Tạm thời tắt check constraint để insert dữ liệu\n")
    f.write("ALTER TABLE SANPHAM NOCHECK CONSTRAINT ALL;\n")
    f.write("ALTER TABLE THUOC NOCHECK CONSTRAINT ALL;\n")

    f.write("DELETE FROM CT_MUAHANG;\nDELETE FROM TOA_THUOC;\n")
    f.write("DELETE FROM THUOC;\nDELETE FROM SANPHAM;\n")

    f.write("-- 1. SANPHAM --\n")
    f.write("".join(sanpham_sql))
    f.write("\n-- 2. THUOC --\n")
    f.write("".join(thuoc_sql))

    f.write("\n-- Bật lại check constraint\n")
    f.write("ALTER TABLE SANPHAM WITH CHECK CHECK CONSTRAINT ALL;\n")
    f.write("ALTER TABLE THUOC WITH CHECK CHECK CONSTRAINT ALL;\n")

print("HOÀN TẤT! Đã tạo file 'insert_master_products.sql'.")
print("File này đã bao gồm lệnh tắt Constraint -> Insert -> Bật lại Constraint.")
```

## 06_Bảng Chi nhánh _ dịch vụ

```python
import random
from itertools import product

SO_CHINHANH = 10
MA_DICHVU = ['DV001', 'DV002', 'DV003']
TRANG_THAI = ["Có cung cấp", "Không cung cấp"]

GIA_MAC_DINH = {
    'DV001': 50000,
    'DV002': 30000,
    'DV003': 50000,
}

def fmt_cn(i):
    """Định dạng Mã Chi Nhánh: CN + 2 số (VD: CN01, CN10)"""
    return f"CN{str(i).zfill(2)}"

def generate_insert_chinhanh_dv():
    """
    Tạo các câu lệnh INSERT INTO cho bảng CHINHANH_DV.
    """

    ma_cn_list = [fmt_cn(i) for i in range(1, SO_CHINHANH + 1)]

    all_combinations = list(product(ma_cn_list, MA_DICHVU))

    sql_lines = []

    sql_lines.append("USE PetCareDB\nGO\n")
    sql_lines.append("DELETE FROM CHINHANH_DV;\n")

    for ma_cn, ma_dv in all_combinations:

        trang_thai = random.choice(TRANG_THAI)
        gia_dv_cn_sql = "NULL"

        if trang_thai == "Có cung cấp" and random.random() < 0.7:
            gia_mac_dinh = GIA_MAC_DINH[ma_dv]
            min_price = int(gia_mac_dinh * 0.8)
            max_price = int(gia_mac_dinh * 1.2)
            gia_dv_cn = round(random.randint(min_price, max_price), -3)
            gia_dv_cn_sql = str(gia_dv_cn)

        sql = (
            f"INSERT INTO CHINHANH_DV (MaCN, MaDichVu, GiaDV_CN, TrangThaiDV) "
            f"VALUES (N'{ma_cn}', N'{ma_dv}', {gia_dv_cn_sql}, N'{trang_thai}');\n"
        )
        sql_lines.append(sql)

    file_name = "insert_chinhanh_dv.sql"
    with open(file_name, "w", encoding="utf-8") as f:
        f.write("".join(sql_lines))

    final_count = len(sql_lines) - 2
    print(f"ĐÃ TẠO FILE {file_name} với {final_count} dòng dữ liệu.")

if __name__ == "__main__":
    generate_insert_chinhanh_dv()
```

## 07_Bảng CT_SPCN

```python
import random

SO_SANPHAM = 500
SO_CHINHANH = 10
SO_CT = 1800

def fmt_sp(i):
    """Format Mã Sản Phẩm: SP00001"""
    return f"SP{str(i).zfill(5)}"

def fmt_cn(i):
    """Format Mã Chi Nhánh: CN01 (2 số) - SỬA QUAN TRỌNG"""
    return f"CN{str(i).zfill(2)}"

gia_sanpham = {
    fmt_sp(i): round(random.randint(20, 1000) * 1000, -3)
    for i in range(1, SO_SANPHAM + 1)
}

with open("insert_ct_spcn.sql", "w", encoding="utf-8") as f:
    sql_lines = []
    sql_lines.append("USE PetCareDB\nGO\n")
    sql_lines.append("DELETE FROM CT_SPCN;\n")
    used_pairs = set()
    while len(used_pairs) < SO_CT:
        cn_id = random.randint(1, SO_CHINHANH)
        sp_id = random.randint(1, SO_SANPHAM)

        cn_str = fmt_cn(cn_id)
        sp_str = fmt_sp(sp_id)

        if (cn_str, sp_str) in used_pairs:
            continue

        used_pairs.add((cn_str, sp_str))

        gia_goc = gia_sanpham[sp_str]

        if random.random() < 0.8:
            giaspcn = gia_goc
        else:
            percent = random.uniform(0.05, 0.15)
            delta = int(gia_goc * percent)

            if random.random() < 0.5:
                giaspcn = gia_goc + delta
            else:
                giaspcn = max(1000, gia_goc - delta)

            giaspcn = round(giaspcn, -3)

        sl_ton = random.randint(0, 200)

        sql = (
            f"INSERT INTO CT_SPCN (MaCN, MaSP, GiaSPCN, SLTonKho) "
            f"VALUES (N'{cn_str}', N'{sp_str}', {giaspcn}, {sl_ton});\n"
        )
        sql_lines.append(sql)

    f.write("".join(sql_lines))

print(f"ĐÃ TẠO FILE insert_ct_spcn.sql VỚI {SO_CT} DÒNG.")
```

```python
import random
from datetime import datetime, timedelta

SO_LS = 2000
SO_NHANVIEN = 1000
SO_CHINHANH = 10

def fmt(prefix, i, length=5):
    """Format cho NV: NV00001"""
    return f"{prefix}{str(i).zfill(length)}"

def fmt_cn(i):
    """Format cho CN: CN01 (2 số)"""
    return f"CN{str(i).zfill(2)}"

def rand_date_before_today():
    today = datetime.now()
    start = datetime(2018, 1, 1)
    delta = today - start
    return (start + timedelta(days=random.randint(0, delta.days))).strftime("%Y-%m-%d")

nv_stt_map = {}

with open("insert_ls_dieudong.sql", "w", encoding="utf-8") as f:
    sql_lines = []

    sql_lines.append("USE PetCareDB\nGO\n")
    sql_lines.append("DELETE FROM LS_DIEUDONG;\n")

    for _ in range(SO_LS):
        nv_id = random.randint(1, SO_NHANVIEN)
        manv = fmt("NV", nv_id)

        current_stt = nv_stt_map.get(manv, 0) + 1
        nv_stt_map[manv] = current_stt

        ngay_dd = rand_date_before_today()

        cn_cu_id = random.randint(1, SO_CHINHANH)
        cn_moi_id = random.randint(1, SO_CHINHANH)

        while cn_moi_id == cn_cu_id:
            cn_moi_id = random.randint(1, SO_CHINHANH)

        cn_cu = fmt_cn(cn_cu_id)
        cn_moi = fmt_cn(cn_moi_id)

        sql = (
            f"INSERT INTO LS_DIEUDONG (STT, MaNV, NgayDieuDong, ChiNhanhCu, ChiNhanhMoi) "
            f"VALUES ({current_stt}, N'{manv}', '{ngay_dd}', N'{cn_cu}', N'{cn_moi}');\n"
        )

        sql_lines.append(sql)

    f.write("".join(sql_lines))

print(f"ĐÃ TẠO FILE insert_ls_dieudong.sql VỚI {SO_LS} DÒNG (STT chuẩn theo từng NV)")
```

## 17_Bảng CT_MUAHANG

```python
import random
import re
import os

FILE_SANPHAM = "insert_master_products.sql"

FILE_LSDV_MUAHANG = "insert_ls_details.sql"

def read_products(filepath):
    """Đọc MaSP và Giá từ file SQL"""
    products = {}
    print(f"Đang đọc giá sản phẩm từ: {filepath}...")

    if not os.path.exists(filepath):
        print(f"LỖI: Không tìm thấy file {filepath}. Hãy sửa lại tên biến FILE_SANPHAM.")
        return {}

    with open(filepath, "r", encoding="utf-8-sig") as f:
        content = f.read()
        matches = re.findall(r"N'(SP\d+)'.*?,\s*(\d+)(\.\d+)?\)", content)
        for m in matches:
            masp = m[0]
            gia = int(float(m[1])) # Lấy giá
            products[masp] = gia

    print(f"-> Đã tìm thấy {len(products)} sản phẩm.")
    return products

def read_transaction_ids(filepath):
    """Đọc MaLSDVMH từ file SQL có sẵn"""
    ids = []
    print(f"Đang đọc mã đơn hàng từ: {filepath}...")

    if not os.path.exists(filepath):
        print(f"LỖI: Không tìm thấy file {filepath}. Hãy sửa lại tên biến FILE_LSDV_MUAHANG.")
        return []

    with open(filepath, "r", encoding="utf-8-sig") as f:
        for line in f:
            if "LS_DVMUAHANG" in line:
                match = re.search(r"VALUES\s*\(\s*N'([^']+)'", line)
                if match:
                    ids.append(match.group(1))

    unique_ids = list(set(ids))

    print(f"-> Đã tìm thấy {len(unique_ids)} mã đơn hàng mua sắm.")
    return unique_ids

def write_batch(f, data_list):
    """Ghi batch insert"""
    BATCH_SIZE = 1000
    if not data_list: return

    total = len(data_list)
    for i in range(0, total, BATCH_SIZE):
        batch = data_list[i : i + BATCH_SIZE]
        values_str = ",\n".join(batch)
        f.write(f"INSERT INTO CT_MUAHANG (MaLSDVMH, MaSP, SoLuongSP, ThanhTienMH) VALUES \n{values_str};\n")

if __name__ == "__main__":
    dict_sp_gia = read_products(FILE_SANPHAM)
    list_ma_lsdv = read_transaction_ids(FILE_LSDV_MUAHANG)

    if not dict_sp_gia or not list_ma_lsdv:
        print("Dừng chương trình do thiếu dữ liệu đầu vào.")
    else:
        all_sp_ids = list(dict_sp_gia.keys())
        data_ct_muahang = []

        print("Đang sinh chi tiết mua hàng...")

        for ma_lsdv in list_ma_lsdv:
            num_items = random.randint(1, 5)
            selected_items = random.sample(all_sp_ids, num_items)

            for ma_sp in selected_items:
                so_luong = random.randint(1, 10)
                don_gia = dict_sp_gia[ma_sp]
                thanh_tien = so_luong * don_gia

                data_ct_muahang.append(f"(N'{ma_lsdv}', N'{ma_sp}', {so_luong}, {thanh_tien})")

        output_file = "insert_ct_muahang_only.sql"
        print(f"Đang ghi file {output_file}...")

        with open(output_file, "w", encoding="utf-8-sig") as f:
            f.write("USE PetCareDB\nGO\n")
            f.write("DELETE FROM CT_MUAHANG;\n")

            write_batch(f, data_ct_muahang)

        print(f"HOÀN TẤT! Đã tạo {len(data_ct_muahang)} dòng chi tiết mua hàng khớp 100% với file cũ.")
```

## 19, 20_Bảng Hóa đơn, chi tiết hóa đơn

```python
import random
from datetime import datetime, timedelta

SO_HOADON = 50000
CT_TOTAL = 100000
SO_NV = 1000
SO_KH = 12000
SO_PET = 18000
SO_CHINHANH = 10

def rand_date():
    start = datetime(2023, 1, 1)
    end = datetime(2024, 12, 31)
    delta = end - start
    return (start + timedelta(days=random.randint(0, delta.days))).strftime("%Y-%m-%d")

def fmt(prefix, i, length=5):
    """Format mã: VD fmt('HD', 1, 6) -> HD000001"""
    return f"{prefix}{str(i).zfill(length)}"

hinhthuc_pay = ["Tiền mặt", "Chuyển khoản", "MOMO", "VNPay"]
trangthai_hd = ["Đã thanh toán", "Chờ thanh toán", "Chưa thanh toán"]

print("Đang tính toán phân bổ dịch vụ...")

all_lsgd_ids = list(range(1, CT_TOTAL + 1))
random.shuffle(all_lsgd_ids)

lsgd_per_invoice = [1] * SO_HOADON
remaining_lsgd = CT_TOTAL - SO_HOADON

for _ in range(remaining_lsgd):
    idx = random.randint(0, SO_HOADON - 1)
    lsgd_per_invoice[idx] += 1

print("Đang tạo file SQL...")

sql_hd = []
sql_ct = []

lsgd_cursor = 0

for i in range(1, SO_HOADON + 1):
    mahd = fmt("HD", i, 8)
    ngay = rand_date()

    ma_cn = f"CN{str(random.randint(1, SO_CHINHANH)).zfill(2)}"

    ma_nv = fmt("NV", random.randint(1, SO_NV), 5)

    ma_kh = fmt("KH", random.randint(1, SO_KH), 5)

    num_services = lsgd_per_invoice[i-1]
    total_bill = 0

    for _ in range(num_services):
        ls_id_val = all_lsgd_ids[lsgd_cursor]
        lsgd_cursor += 1

        ma_lsgd = fmt("LSDV", ls_id_val, 6)

        phi_dv = random.choice([30000, 50000, 150000, 200000, 500000])
        total_bill += phi_dv

        pet_val = "NULL"
        if random.random() > 0.3:
            pet_val = f"N'{fmt('PET', random.randint(1, SO_PET), 5)}'"

        sql_ct.append(f"INSERT INTO CT_HOADON VALUES (N'{mahd}', N'{ma_lsgd}', {phi_dv}, {pet_val});\n")

    cong_loyalty = total_bill // 50000

    tien_thanh_toan = total_bill
    if random.random() > 0.8:
        giam = int(total_bill * random.choice([0.05, 0.1]))
        tien_thanh_toan = total_bill - giam

    tien_truoc_km = total_bill
    hinh_thuc = random.choice(hinhthuc_pay)
    trang_thai = random.choice(trangthai_hd)

    row_hd = (f"INSERT INTO HOADON VALUES ("
              f"N'{mahd}', '{ngay}', N'{ma_nv}', {tien_truoc_km}, {tien_thanh_toan}, "
              f"N'{hinh_thuc}', N'{trang_thai}', {cong_loyalty}, N'{ma_kh}', N'{ma_cn}');\n")
    sql_hd.append(row_hd)

print(f"Đang ghi {SO_HOADON} dòng vào insert_hoadon.sql...")
with open("insert_hoadon.sql", "w", encoding="utf-8") as f:
    f.write("USE PetCareDB\nGO\n")
    f.write("".join(sql_hd))

print(f"Đang ghi {len(sql_ct)} dòng vào insert_ct_hoadon.sql...")
with open("insert_ct_hoadon.sql", "w", encoding="utf-8") as f:
    f.write("USE PetCareDB\nGO\n")
    f.write("".join(sql_ct))

print("HOÀN TẤT! Đã sinh dữ liệu khớp với ràng buộc.")
```

## 21_Bảng Chi tiết khuyến mãi

```python
import random

SO_HOADON = 50000

KHUYENMAI_INFO = {
    "KM001": {"LoaiKM": "HVCoBan", "GiaKM": 3},
    "KM002": {"LoaiKM": "HVThanThiet", "GiaKM": 5},
    "KM003": {"LoaiKM": "HVVIP", "GiaKM": 10},
    "KM004": {"LoaiKM": "Loyalty", "GiaKM": 1000},
}

def fmt(prefix, i, length=8):
    """Định dạng mã với prefix và padding 8 số (cho MaHD)."""
    return f"{prefix}{str(i).zfill(length)}"

def generate_insert_ct_khuyenmai():
    print("Đang tính toán phân bổ khuyến mãi...")

    hd_with_2_km = set(random.sample(range(1, SO_HOADON + 1), k=20000))
    used_mahd_makm = set()
    sql_lines = []
    sql_lines.append("USE PetCareDB\nGO\n")

    print("Đang tạo SQL...")
    for i in range(1, SO_HOADON + 1):
        ma_hd = fmt("HD", i, 8)
        num_km = 2 if i in hd_with_2_km else 1
        available_kms = list(KHUYENMAI_INFO.keys())
        selected_kms = random.sample(available_kms, k=min(num_km, len(available_kms)))

        tien_truoc_km_mock = random.randint(100000, 5000000)

        for ma_km in selected_kms:
            if (ma_hd, ma_km) in used_mahd_makm:
                continue

            km_info = KHUYENMAI_INFO[ma_km]
            loai_km = km_info["LoaiKM"]
            gia_km = km_info["GiaKM"]

            so_luong_dung = 0
            tien_km = 0

            if loai_km == "Loyalty":
                so_luong_dung = random.randint(10, 200)
                tien_km = so_luong_dung * gia_km

                if tien_km > tien_truoc_km_mock:
                    tien_km = int(tien_truoc_km_mock * 0.5)
                    so_luong_dung = tien_km // 1000
            else:
                so_luong_dung = 1
                tien_km = int(tien_truoc_km_mock * (gia_km / 100.0))

            used_mahd_makm.add((ma_hd, ma_km))

            sql = (
                f"INSERT INTO CT_KHUYENMAI (MaHD, MaKM, SoLuongDung, TienKM) "
                f"VALUES (N'{ma_hd}', N'{ma_km}', {so_luong_dung}, {tien_km});\n"
            )
            sql_lines.append(sql)

    file_name = "insert_ct_khuyenmai.sql"
    with open(file_name, "w", encoding="utf-8") as f:
        f.write("".join(sql_lines))

    final_count = len(sql_lines) - 1
    print(f"ĐÃ TẠO FILE {file_name} với {final_count} dòng dữ liệu.")

if __name__ == "__main__":
    generate_insert_ct_khuyenmai()
```

## 22_Bảng Đánh giá

```python
import random

SO_KH = 12000
SO_DANHGIA = 70000
KH_DANHGIA_RATE = 0.35

MA_DICHVU = ['DV001', 'DV002', 'DV003']

DIEM_DANHGIA = [1, 2, 3, 4, 4, 5, 5, 5, 5, 5]

BINHLUAN_MAU = [
    "Dịch vụ rất tốt, nhân viên thân thiện.",
    "Khám bệnh nhanh chóng và chuyên nghiệp.",
    "Rất hài lòng với sự tư vấn của nhân viên.",
    "Chưa thực sự hài lòng về tốc độ phục vụ.",
    "Mức độ hài lòng: Bình thường.",
    "Tuyệt vời! Bé nhà mình rất thích bác sĩ.",
    "Giá cả hợp lý, sẽ quay lại.",
    "Cần cải thiện khâu vệ sinh chuồng trại.",
    "Bác sĩ rất tận tâm.",
    None,
    None,
    None,
    None
]

def fmt(prefix, i, length=5):
    """Định dạng mã với prefix và padding."""
    return f"{prefix}{str(i).zfill(length)}"

def generate_insert_danhgia():
    print("Đang tính toán phân bổ khách hàng...")
    so_luong_kh_active = int(SO_KH * KH_DANHGIA_RATE)

    active_kh_ids = random.sample(range(1, SO_KH + 1), so_luong_kh_active)

    selected_kh_ids = random.choices(active_kh_ids, k=SO_DANHGIA)

    sql_lines = []
    print(f"Đang tạo {SO_DANHGIA} câu lệnh INSERT...")

    for i in range(1, SO_DANHGIA + 1):
        ma_dg = fmt("DG", i, 7)

        diem_dv = random.choice(DIEM_DANHGIA)
        diem_nv = random.choice(DIEM_DANHGIA)
        muc_do_hl = random.choice(DIEM_DANHGIA)

        binh_luan_raw = random.choice(BINHLUAN_MAU)
        if binh_luan_raw is None:
            binh_luan_sql = "NULL"
        else:
            clean_binh_luan = binh_luan_raw.replace("'", "''")
            binh_luan_sql = f"N'{clean_binh_luan}'"

        ma_dv = random.choice(MA_DICHVU)

        kh_id = selected_kh_ids[i-1]
        ma_kh = fmt("KH", kh_id, 5)

        sql = (
            f"INSERT INTO DANHGIA (MaDG, DiemDV, DiemNV, MucDoHaiLong, BinhLuan, MaDV, MaKH) "
            f"VALUES (N'{ma_dg}', {diem_dv}, {diem_nv}, {muc_do_hl}, "
            f"{binh_luan_sql}, N'{ma_dv}', N'{ma_kh}');\n"
        )
        sql_lines.append(sql)

    file_name = "insert_danhgia.sql"
    with open(file_name, "w", encoding="utf-8") as f:
        f.write("USE PetCareDB\nGO\n")
        f.write("".join(sql_lines))

    print(f"HOÀN TẤT! Đã tạo file {file_name} với {SO_DANHGIA} dòng.")

if __name__ == "__main__":
    generate_insert_danhgia()
```
